```python
import os
import shutil
import numpy as np

def mylistdir(directory, end=''):
    filelist = os.listdir(directory)
    return [x for x in filelist if not x.startswith('.') and x.endswith(f'{end}')]

def timeround5(dt):
    minute_5 = int(5*np.floor(dt.minute/5))
    d2 = dt.replace(minute=minute_5)
    dH, dM = d2.strftime("%H"), d2.strftime("%M")
    return f'{dH}{dM}'


def get_files(path, date_f):
    date_path = os.path.join(path, 'env_from_pi', date_f)
    minute_folders = sorted(mylistdir(date_path))
    for m in minute_folders:
        all_files = sorted(mylistdir(os.path.join(date_path, m), '.json'))
        for f in all_files:
            day_store, minute = f.split(' ')
            min_actual = minute.split('_')[0]
            min_store = timeround5(datetime.strptime(min_actual, '%H%M'))
            f_to_move = os.path.join(date_path, m, f)
            target_dir = os.path.join(path, 'env_params', day_store, min_store)
            if not os.path.exists(target_dir):
                os.makedirs(target_dir)
            try:
                dest = shutil.move(f_to_move, target_dir)
            except Exception as e:
                print(f'Error moving: {e}')
        


####### TO RUN ####   
root_path = '/Users/maggie/Desktop/HPD_w_moved/H1-red/'
hubs = ['RS1', 'RS2', 'RS3', 'RS4', 'RS4']    

for hub in hubs:  
    print(hub)
    folders = sorted(mylistdir(os.path.join(root_path, hub, 'env_params')))
    for date_f in folders:
        get_files(os.path.join(root_path, hub), date_f)
```
